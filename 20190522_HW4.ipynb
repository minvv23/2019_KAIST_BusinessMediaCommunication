(a)  
데이터 사용량 $x$에 대해서 통신사 이용자의 효용(utility)은 $U(x)$로, 가격은 $P(x)$로 표현될 수 있다.  
각각의 수식은 이하와 같이 정의된다.

#### $U(x) = \sigma logx$  
#### $P(x) = g + hx$
($g$ is flat rate fee and $h$ is usage fee)

이 때 소비자의 효용 극대화 지점인 $x^*$는 net utility를 0으로 하는 수준의 $\frac{d}{dx}\{U(x)-P(x)\}=0$에서 결정된다.  
구체적인 계산 과정은 다음과 같다.

#### $U(x)-p(x) = \sigma logx - (g+hx) = 0$

#### $\frac{d}{dx}\{U(x)-P(x)\} = \frac{\sigma}{x^*}-h=0$

#### $x^*=\frac{\sigma}{h}$

해당 $x^*$를 다시 원식에 대입하면, 아래의 과정을 통해 g와 h의 관계식을 도출할 수 있다.

#### $\sigma log(\frac{\sigma}{n}) = g+h * \frac{\sigma}{h}$

#### $g = \sigma\{ log(\frac{\sigma}{h})-1\}$

이 때 $\sigma = \{50, 100, 150, 200\}$의 네 가지 경우에 대해 시각화하면 아래와 같은 관계가 나타난다.

```python
import matplotlib.pyplot as plt, numpy as np, pandas as pd
def g_generator(sigma, h) :
    return sigma*(np.log(sigma/h)-1)

h = np.arange(.00001,100,0.003)
g = [g_generator(x, h) for x in [50,100,150,200]]

for each_g in g :
    plt.plot(h, each_g)
plt.ylim(0,500)
plt.title('h-g visualization')
plt.ylabel('flat rate fee (g)')
plt.xlabel('usage fee (h)')
plt.legend(['50','100','150','200'])
```

(b)  
본 모델에 의하면 주어진 $\sigma$에 대해서 가격인 $P(x^*) = g+hx^* = g+\sigma$를 책정할 때, $g$를 높일수록 기업의 이익이 극대화된다. 이 때 위의 도표에서 볼 수 있듯 g와 h는 반비례 관계에 있으며, $\sigma$값이 커짐에 따라 ($h>0$에 대해) flat rate fee $(g)$의 상한선이 확대된다는 것을 알 수 있다.

(c)  
마지막으로 $\sigma$는 $U(x) = \sigma logx$의 식에서 볼 수 있듯, 동일한 데이터 사용량($x$) 하에서 느끼는 효용의 계수, 혹은 가중치이다. 그렇기에 어떠한 서비스에서 제공하는 컨텐츠의 (정성적인) 질을 높인다거나, 전반적인 데이터 이용 속도를 높이는 등의 policy를 도입할 경우 같은 양의 데이터를 이용하더라도 사용자의 만족감은 보다 높아질 수 있을 것이다. 또한 그 경우, 앞서 (a)와 (b)에서 관찰한 바와 같이 소비자에게 더 높은 flat rate fee ($g$)를 부과할 수 있게 된다.
